## ML Modles

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,f1_score,recall_score
from sklearn.model_selection import train_test_split
import time
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, roc_auc_score, classification_report
```

```python
def function(x_train, y_train, x_test, y_test):
    print("\n")
    r = int(input("enter random state= "))
    # gb = GradientBoostingClassifier(random_state=r)
    # start_time1 = time.time()
    # gb.fit(x_train, y_train)
    # total_time1 = time.time() - start_time1

    gb = GradientBoostingClassifier(random_state=r)
    knn = KNeighborsClassifier()
    gaus = GaussianNB()
    lr =LogisticRegression(random_state=r)
    svc = SVC(random_state=r)
    dt = DecisionTreeClassifier(random_state=r)
    rf = RandomForestClassifier(random_state=r)
    model = [gb, knn, gaus, lr, svc, dt, rf]

    start_time = []
    total_time = []
    y_pred = []
    accuracy = []
    precision = []
    f1 = []
    recall = []
    confusion = []
    fpr_list = []
    tpr_list = []
    auc_list = []
    class_report_list = []

    for i in range(len(model)):
        start = time.time()
        model[i].fit(x_train, y_train)
        end = time.time()
        pred = model[i].predict(x_test)
        #start_time.append(start)
        total_time.append(end - start)
        y_pred.append(pred)
        accuracy.append(accuracy_score(y_test, pred))
        precision.append(precision_score(y_test, pred))
        f1.append(f1_score(y_test, pred))
        recall.append(recall_score(y_test, pred))
        confusion.append(confusion_matrix(y_test, pred))
        fpr, tpr, _ = roc_curve(y_test, pred)
        fpr_list.append(fpr)
        tpr_list.append(tpr)
        auc_list.append(roc_auc_score(y_test, pred))
        class_report_list.append(classification_report(y_test, pred))

    for i in range(len(model)):
        print(model[i])
        print("Training Time:", total_time[i])
        print("Accuracy:", accuracy[i])
        print("Precision:", precision[i])
        print("Recall:", recall[i])
        print("F1 Score:", f1[i])
        print("\nConfusion Matrix:\n", confusion[i])
        print("\nClassification Report:\n", class_report_list[i])

        plt.figure(figsize=(6, 4))
        sns.heatmap(confusion[i], annot=True, fmt='g', cmap='Blues', xticklabels=rf.classes_, yticklabels=rf.classes_)
        plt.xlabel('Predicted labels', fontsize=14, fontname='Times New Roman')
        plt.ylabel('True labels', fontsize=14, fontname='Times New Roman')
        plt.title('Confusion Matrix', fontsize=14, fontname='Times New Roman')
        plt.show()

        plt.figure()
        plt.plot(fpr_list[i], tpr_list[i], color='darkorange', label=f'ROC curve (AUC = {auc_list[i]:.4f})')
        plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Receiver Operating Characteristic (ROC) curve')
        plt.legend(loc='lower right')
        plt.grid(True)
        plt.show()
```
